```python
from sklearn.metrics import roc_curve,roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier

# Assuming X is our feature matrix and Y is the target vector
# Binarize the output (Making bins for the outpus)
Y_bin = label_binarize(Y, classes=[0, 1, 2, 3, 4])  # Adjusting the classes
n_classes = Y_bin.shape[1]

clf = OneVsRestClassifier(DecisionTreeClassifier())
clf.fit(X, Y_bin)

# Compute the ROC curve and ROC AUC score for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(Y_bin[:, i], clf.predict_proba(X)[:, i])
    roc_auc[i] = roc_auc_score(Y_bin[:, i], clf.predict_proba(X)[:, i])

# Plotting the ROC curves for each class
import matplotlib.pyplot as plt

plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (area = {roc_auc[i]:.2f})')

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.model_selection import train_test_split

# Assuming X is our feature matrix and Y is our target variable

Y_bin = label_binarize(Y, classes=[0, 1, 2,3,4])  
n_classes = Y_bin.shape[1]
# Split the data into train and test sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y_bin, test_size=0.3, random_state=42)


clf = OneVsRestClassifier(DecisionTreeClassifier())
clf.fit(X_train, Y_train)

# Predicting the probabilities
Y_score = clf.predict_proba(X_test)

# Compute ROC curve and ROC AUC score for each class
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(Y_test[:, i], Y_score[:, i])
    roc_auc[i] = roc_auc_score(Y_test[:, i], Y_score[:, i])

# Compute macro-average ROC curve and ROC AUC
fpr["macro"], tpr["macro"], _ = roc_curve(Y_test.ravel(), Y_score.ravel())
roc_auc["macro"] = roc_auc_score(Y_test, Y_score, average="macro")

# Compute micro-average ROC curve and ROC AUC
fpr["micro"], tpr["micro"], _ = roc_curve(Y_test.ravel(), Y_score.ravel())
roc_auc["micro"] = roc_auc_score(Y_test, Y_score, average="micro")

# Plotting the ROC curves for each class and the macro-average ROC curve
import matplotlib.pyplot as plt

plt.figure()
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} (area = {roc_auc[i]:.2f})')

plt.plot(fpr["macro"], tpr["macro"], label='macro-average ROC curve (area = {0:0.2f})'
                                           ''.format(roc_auc["macro"]),
         color='navy', linestyle=':', linewidth=4)

plt.plot(fpr["micro"], tpr["micro"], label='micro-average ROC curve (area = {0:0.2f})'
                                           ''.format(roc_auc["micro"]),
         color='deeppink', linestyle=':', linewidth=4)

plt.plot([0, 1], [0, 1], 'k--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curves')
plt.legend(loc="lower right")
plt.show()
```

```python
##Hyperparameter in decision trees are important features which controls the behavior and the structure of the model during the training phase.

#Criteria : The quality of the split in the decision tree is measured by this function called criteria.

#The criteria support two types such as gini (Gini impurity) and entropy (information gain)

#Gini index: The measure that divides the probability distributions of the target attribute's values is known as the Gini impurity or Gini index.
#The node is split in a way that produces the least amount of impurity possible.

#Information gain: This impurity measure divides a node in a way that maximises information gain by applying the entropy measure.

#max_depth: As the name is suggesting, max_depth hyperparameter  will be controlling the maximum depth for which the decision tree is allowed to grow.
    
#The tree can identify more intricate patterns in the training data when the max_depth is deeper, which may help to lower the training error.
#Max_depth should not be set too high though, as this can cause overfitting, in which the model learns to ignore the noise in the training set. 

#It is very important to tune max_depth carefully to find the right balance between model complexity and generalization performance. 

#the input option for max_depth can be a positive integer or â€˜Noneâ€™ that indicates no maximum depth limit.

#min_samples_split: The min_sample_split hyperparameter defines the minimal number of samples that are needed to split a node. 

#Note that the min_samples_split function acts as a threshold in a decision tree; 
#if a node has fewer samples than the threshold, the node will not be split and will instead become a leaf node.

#min_samples_leaf hyperparameter specifies the bare minimum of samples that must exist at a leaf node.
#It serves as a cutoff point that marks a node as a leaf and ends the splitting process.

#The input option for min_samples_leaf can either be an integer or float.

#max_features: We may regulate the number of features taken into account while determining the optimal split in the 
#decision tree by using the max_features hyperparameter. 
    
#It can either define an exact number of features to consider at each split or as a percentage that represents the proportion of features to consider.

#The input options can be an integer, float, auto, sqrt, log2.

#auto â€“ It allows the decision tree algorithm to consider all the features for each split.

#sqrt â€“ It allows the algorithm to consider only the square root of the total number of features for each split

#log2 â€“ It allows the algorithm to consider the logarithm base 2 of a total number of features for each split.

#min_weight_fraction_leaf: This hyperparameter is used to regulate the topology of the tree according to the weights given to each sample. 
  
#This hyperparameter determines the minimum part of input samples required at a leaf node. 
```
